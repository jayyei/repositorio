# Comprensi√≥n de listas

```python
# Metodo tradicional 
lista = []

for letra in 'casa':
    lista.append(letra)
    
print(lista)
```

```python
# Metodo con compresion de listas
lista = [letra for letra in 'casa']
print(lista)
```

```python
# Metodo tradicional
lista = []
for numero in range(0,11):
    lista.append(numero**2)
    
print(lista)
```

```python
# metod con compresion de listas
lista = [numero**2 for numero in range(0,11)]
print(lista)
```

```python
# Metodo tradicional
lista = []
for numero in range(0,11):
    if numero % 2 == 0:
        lista.append(numero)
        
print(lista)
```

```python
lista = [ numero for numero in range(0,11) if numero %2 == 0 ]
print(lista)
```

```python
# Metodo tradicional
lista = []
for numero in range(0,11):
    lista.append(numero**2)

pares = []
for numero in lista:
    if numero % 2 == 0:
        pares.append(numero)
print(pares)        
```
